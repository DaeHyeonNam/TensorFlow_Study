```python
import tensorflow as tf

from tensorflow.examples.tutorials.mnist import input_data

mnist= input_data.read_data_sets("/tmp/data/", one_hot = True)

x= tf.placeholder(tf.float32, shape= [None, 784])
y= tf.placeholder(tf.float32, shape= [None, 10])
```

```python
def CNNFilter(x):
    x_image = tf.reshape(x, [-1,28,28,1])
    #28*28*1 -> 28*28*32
    W_CNN_1 = tf.Variable(tf.truncated_normal(shape= [5,5,1,32], stddev=5e-2))
    b_CNN_1 = tf.Variable(tf.constant(0.1, shape = [32]))
    h_CNN_1 = tf.nn.relu(tf.nn.conv2d(x_image,W_CNN_1, strides= [1,1,1,1] ,padding = 'SAME')+ b_CNN_1)
    #28*28*32 -> 14*14*32
    h_pool1 = tf.nn.max_pool(h_CNN_1, ksize= [1,2,2,1], strides= [1,2,2,1], padding='SAME')
    #14*14*32 -> 14*14*64
    W_CNN_2 = tf.Variable(tf.truncated_normal(shape= [5,5,32,64], stddev=5e-2))
    b_CNN_2 = tf.Variable(tf.constant(0.1, shape= [64]))
    h_CNN_2 = tf.nn.relu(tf.nn.conv2d(h_pool1,W_CNN_2, strides = [1,1,1,1], padding= 'SAME')+ b_CNN_2)
    #14*14*64 -> 7*7*64
    h_pool2 = tf.nn.max_pool(h_CNN_2, ksize= [1,2,2,1], strides= [1,2,2,1], padding = 'SAME')
    
    return h_pool2
```

```python
def ANNFilter(x):
    W1 = tf.Variable(tf.random_normal(shape= [7*7*64, 1024], stddev=5e-2))
    b1 = tf.Variable(tf.constant(0.1, shape= [1024]))
    h_pool2_flat = tf.reshape(x, [-1,7*7*64])
    h1 = tf.nn.relu(tf.matmul(h_pool2_flat,W1)+ b1)
    
    W_output = tf.Variable(tf.random_normal(shape= [1024, 10], stddev=5e-2))
    b_output = tf.Variable(tf.constant(0.1, shape= [10]))
    h_output = tf.matmul(h1, W_output)+ b_output
    
    return h_output
```

```python
CNN_model = CNNFilter(x)
output_model = ANNFilter(CNN_model)

y_pred = tf.nn.softmax(output_model)

cost_function = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels = y, logits= output_model))
optimizer = tf.train.AdamOptimizer(1e-4)
train_step = optimizer.minimize(cost_function)

correct_prediction = tf.equal(tf.argmax(y_pred, 1), tf.argmax(y,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for i in range(10000):
        batch =mnist.train.next_batch(50)
        
        if (i+1)%100 ==0:
            train_accuracy = accuracy.eval(feed_dict={x:batch[0], y:batch[1]})
            print("반복: {}, 정확도: {}".format(i, train_accuracy))
        sess.run(train_step, feed_dict={x:batch[0], y:batch[1]})
    
    
```
